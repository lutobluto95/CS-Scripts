```python
import pandas as pd
import io
from google.colab import files
```

```python
data = files.upload()
```

```python
y = pd.read_excel(io.BytesIO(data[f'file.xlsx']))
y.shape
```

```python
mentions_col = []

for i in range(y.shape[0]):
    listm = []
    mentions = y.iloc[[i], 6:]
    sum = int(mentions.sum(axis=1))
    if sum > 0:    
        for j, col in enumerate(mentions.columns):
            value = mentions.iloc[0, j]
            if value > 0: listm.append(f'{col}-1')
    else: listm.append('No Mentions')
    final_value = ', '.join(listm)
    mentions_col.append(final_value)
```

```python
y['MentionsValue'] = mentions_col
```

```python
y.to_excel('file_structured.xlsx', index=False)
files.download('file_structured.xlsx')
```
