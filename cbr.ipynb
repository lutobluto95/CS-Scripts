```python
prod = data[['metricForDayUTCMonth','metricForDayUTCYear','top10Keywords']]
prod.head()
```

```python
p = pd.json_normalize(prod.top10Keywords.apply(json.loads))
```

```python
pp = pd.concat([prod,p],axis=1)
pp.drop(columns='top10Keywords',inplace=True)
pp.head()
```

```python
ppp = pp.melt(id_vars=['metricForDayUTCMonth','metricForDayUTCYear'],var_name='Keywords',value_name='Value').dropna(subset='Value')
ppp.head()
```

```python
ppp[['Keywords','Type']] = ppp.Keywords.str.split('.',expand=True)
ppp.tail(20)
```

```python
ppp.shape[0]
```

```python
lookup = ppp[ppp['Value'] == 'Products'][['Keywords','Value']]
lookup.drop_duplicates(subset='Keywords',inplace=True)
lookup
```

```python
products = ppp.merge(lookup, on='Keywords')
products = products[products.Type=='numOccurrences']
products
```

```python
products.info()
```

```python
products.columns = ['month','year','keywords','occurences','type','type2']
products['day'] = 1
products['date'] = pd.to_datetime((products.year*10000+products.month*100+products.day).apply(str),format='%Y%m%d')
products.head()
```

```python
products.occurences = products.occurences.astype('int16')
products.head()
```

```python
productsd = products.groupby(['date','keywords'])[['occurences']].sum()
productsd.reset_index(inplace=True)
productsd
```

```python
productsd = productsd.sort_values(['date','occurences'],ascending=[True,False])
productsd
```
